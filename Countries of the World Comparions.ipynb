```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 



df = pd.read_csv('C:\\Users\\Square Bear\\Documents\\Thinkful\\Datasets\\countries of the world.csv')

df = df.dropna()

df['Literacy (%)'] = df['Literacy (%)'].str.replace(',','.')

df['Literacy (%)'] = df['Literacy (%)'].apply(float)

plt.scatter(x = df['Population'], y = df['Literacy (%)'], color='brown')
plt.title('Countries of the World Comparison')
plt.xlabel('Population')
plt.ylabel('Literacy (%)')
plt.show()
```

### Population has no correlation to literacy. No matter what size of the population, literacy varies. 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 



df = pd.read_csv('C:\\Users\\Square Bear\\Documents\\Thinkful\\Datasets\\countries of the world.csv')

df = df.dropna()

df['Literacy (%)'] = df['Literacy (%)'].str.replace(',','.')

df['Literacy (%)'] = df['Literacy (%)'].apply(float)

df['Phones (per 1000)'] = df['Phones (per 1000)'].str.replace(',','.')

df['Phones (per 1000)'] = df['Phones (per 1000)'].apply(float)

plt.scatter(x = df['Phones (per 1000)'], y = df['Literacy (%)'], color='red')
plt.title('Countries of the World Comparison')
plt.xlabel('Phones')
plt.ylabel('Literacy (%)')
plt.show()
```

### As the number of phones increase, the literacy increases. There are instances where there is high literacy and no phones so maybe the better statement is as literacy increases, the phones serve as an option rather than a need. 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 



df = pd.read_csv('C:\\Users\\Square Bear\\Documents\\Thinkful\\Datasets\\countries of the world.csv')

df = df.dropna()

df['Literacy (%)'] = df['Literacy (%)'].str.replace(',','.')

df['Literacy (%)'] = df['Literacy (%)'].apply(float)

plt.scatter(x = df['GDP ($ per capita)'], y = df['Literacy (%)'], color='blue')
plt.title('Countries of the World Comparison')
plt.xlabel('GDP ($ per capita)')
plt.ylabel('Literacy (%)')
plt.show()
```
