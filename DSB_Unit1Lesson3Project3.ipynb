```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


suicides = pd.read_csv('C:\\Users\\Square Bear\\Documents\\Thinkful\\Datasets\\master.csv', thousands=',')


suicides.head(3)







```

```python
suicides2 = suicides.groupby(['country','year'])['suicides_no'].sum()
suicides3 = suicides.groupby(['country','year'])['population'].sum()
suicides4 = suicides.groupby(['country','year'])[' gdp_for_year ($) '].sum()
suicides5 = suicides.groupby(['country','year'])['gdp_per_capita ($)'].sum()
suicides6 = suicides.groupby(['country','year'])['suicides/100k pop'].mean()
suicides7 = suicides.groupby(['country','year'])['suicides_no', 'population'].sum()

print(suicides7.head())
```

```python
suicides2 = suicides2.reset_index()
```

```python
usa2 = suicides2[(suicides2.country == 'United States')]
print(usa2.head())
```

```python
usa2 = usa2.reset_index()
```

```python



plt.figure(figsize = (20, 20))
plt.subplot(2,2,1)
plot_1 = sns.pointplot(orient='h', data=usa2['suicides_no'])


plt.subplot(2,2,2)
plot_2 = sns.boxplot(orient = 'h', data=usa2['suicides_no'])


plt.subplot(2,2,3)
plot_3 = sns.countplot(y='suicides_no', orient ='h', data=usa2, palette='colorblind')


plt.subplot(2,2,4)
plot_4 = sns.violinplot(orient = 'h', data=usa2['suicides_no'])



```
