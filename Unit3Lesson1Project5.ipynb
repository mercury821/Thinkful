```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics

df = pd.DataFrame(columns = ['Name', 'Age']) 

df['Name'] = ['Greg', 'Marcia', 'Peter', 'Jan', 'Bobby', 'Oliver', 'Cindy']
df['Age'] = [14, 12, 11, 10, 8, 8, 6]

print(df)
print('\n')
print('The mean is: ', np.mean(df['Age']))
print('The median is: ', np.median(df['Age']))
print('The mode is 8.')
print('The standard deviation is: ', np.std(df['Age']))
print('The variance is: ', np.var(df['Age']))
print('The standard error is: ', np.std(df['Age'] ,ddof=1) / np.sqrt(len(df['Age'])))
print('\n')


print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('\n')

print('If I had to choose one measure of central tendancy, that would be the median')
print('which would provide the midpoint of the data')
print('\n')
print('If I had to choose one measure of variance, I would choose the standard deviation')
print('which would provide the approximate range of my data as well as the distance between data points.')
print('\n')


print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('\n')

df.replace: df['Age'].replace(6,7, inplace = True)
      
print(df)   
print('\n')
print('The mean with a new birthday for Cindy is: ', np.mean(df['Age']))
print('The median is: ', np.median(df['Age']))
print('The mode is 8.')
print('The standard deviation is: ', np.std(df['Age']))
print('The variance is: ', np.var(df['Age']))
print('The standard error is: ', np.std(df['Age'] ,ddof=1) / np.sqrt(len(df['Age'])))

print('\n')
print('The mean, mode, variance, standard deviation, and standard error have change.')
print('I would use the median. It still represents the data well.')
print('\n')

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print('\n')

df.replace: df['Name'].replace('Oliver','Jessica', inplace = True)

df.loc[(df['Age'] == 8)[::-1].idxmax(), 'Age'] = 1

print(df)  
print('\n')
print('The mean with Jessica is: ', np.mean(df['Age']))
print('The median is: ', np.median(df['Age']))
print('There is no mode.')
print('The standard deviation is: ', np.std(df['Age']))
print('The variance is: ', np.var(df['Age']))
print('The standard error is: ', np.std(df['Age'] ,ddof=1) / np.sqrt(len(df['Age'])))
print('\n')

print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')


print('\n')
Bbunch = [ 17, 20, 23]
print('The 50th Anniversary of the Brady Bunch had a mean of %.2f percent and a median of %.1f percent.' % (np.mean(Bbunch), np.median(Bbunch)))


```
